5. Any outliers to remove?

```python
# Identification of outliers using the IQR method
def identify_outliers_iqr(df, column, threshold=1.5):
    Q1 = customers[column].quantile(0.25)
    Q3 = customers[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR

    outliers = customers[(df[column] < lower_bound) | (customers[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

for col in spend_cols:
    col_outliers, col_lower, col_upper = identify_outliers_iqr(customers, col)
    print(f"IQR Method: {col} outliers detected: {len(col_outliers)} (bounds: {col_lower:.2f} to {col_upper:.2f})")
```

```python
# Remove outliers from continuous variables using the IQR method (1.5 × IQR fence)
def remove_outliers_iqr(df):
    Q1 = customers[spend_cols].quantile(0.25)
    Q3 = customers[spend_cols].quantile(0.75)
    IQR = Q3 - Q1
    mask = ~((customers[spend_cols] < (Q1 - 1.5 * IQR)) | (customers[spend_cols] > (Q3 + 1.5 * IQR))).any(axis=1)
    return customers[mask].copy()

customers_outliers_handled = remove_outliers_iqr(customers)

print(f"Original shape: {df.shape}")
print(f"After outlier removal shape: {customers_outliers_handled.shape}")
print(f"Rows removed: {df.shape[0] - customers_outliers_handled.shape[0]}")
display(customers_outliers_handled.describe().T)
```

# Challenge 4 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python
%pip install kneed
```

```python
# Install first: pip install kneed

from kneed import KneeLocator
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

def find_elbow(data, max_k=10):
    inertia = []
    for k in range(1, max_k + 1):
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(data)
        inertia.append(kmeans.inertia_)
    
    knee_locator = KneeLocator(
        range(1, max_k + 1), inertia,
        curve="convex", direction="decreasing"
    )
    optimal_k = knee_locator.knee
    
    plt.plot(range(1, max_k + 1), inertia, marker='o')
    plt.axvline(x=optimal_k, linestyle='--', label=f'Optimal k={optimal_k}')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Inertia')
    plt.title('Elbow Method')
    plt.legend()
    plt.show()
    
    return optimal_k
```

```python
find_elbow(customers_scale[spend_cols])
```

```python
from sklearn.metrics import silhouette_score

silhouettes = []
for k in range(2, 11):  # silhouette needs k>=2
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(customers_scale[spend_cols])
    silhouettes.append(silhouette_score(customers_scale[spend_cols], labels))

plt.plot(range(2, 11), silhouettes, marker='o')
plt.xlabel('k'); plt.ylabel('Silhouette score')
plt.show()
```

```python
import pandas as pd
pd.Series(labels).value_counts()
```

The elbows indicates k = 4, but the silhouette score indicates k = 2, so we'll proceed with k = 2.
